```python
import soundfile as sf
import sounddevice as sd

import numpy as np
from pydub import AudioSegment
from ctypes import cast, POINTER
from comtypes import CLSCTX_ALL
from pycaw.pycaw import AudioUtilities, IAudioEndpointVolume
import math

```

```python

# Get default audio device using PyCAW
devices = AudioUtilities.GetSpeakers()
interface = devices.Activate(
    IAudioEndpointVolume._iid_, CLSCTX_ALL, None)
volume = cast(interface, POINTER(IAudioEndpointVolume))
# Get current volume 
currentVolumeDb = volume.GetMasterVolumeLevel()
print(currentVolumeDb) #-->-65.25,-23.654823303222656, -13.61940860748291,-7.626824855804443,-3.339998245239258,0.0
#volume.SetMasterVolumeLevel(currentVolumeDb +1, None)
```

```python
filename = 'audio'
path = 'Audio_clips/' + filename + '.wav' # path from my computer 
audio_file, fs = sf.read(path, dtype='float32')  
```

```python
def play(audio_file):
    sd.play(audio_file, 44100)
```

```python
def stop():
    sd.stop()
```

```python
play(audio_file)
```

```python
stop() #stop audio
```

```python
def volume_20():
    devices = AudioUtilities.GetSpeakers()
    interface = devices.Activate(
        IAudioEndpointVolume._iid_, CLSCTX_ALL, None)
    volume = cast(interface, POINTER(IAudioEndpointVolume))
    setup_20 = -23.654823303222656
    volume.SetMasterVolumeLevel(setup_20, None)
```
